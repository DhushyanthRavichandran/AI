#identifing outliers


1.   z score   (x-xmean)/std deviation
2.   IQR inter quartile range ->IQR approach to finding the outliers is the most commonly used and most trusted approach used in research feilds



```python
from scipy import stats
```

```python
data
```

```python
hp_zscore=abs(stats.zscore(data['normalized-losses']))
hp_zscore
```

```python
threshold =hp_zscore.max()
threshold
data['normalized-losses'].min()
```

```python
index_value=np.where(hp_zscore>=threshold)
data.iloc[index_value]['normalized-losses']
```

```python
data.drop(index=data.iloc[index_value]['normalized-losses'].index,inplace=True)
data
```
