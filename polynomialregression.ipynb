```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Real%20estate.csv')
data
```

```python
data.info()
```

```python
data.corr()
```

```python
plt.subplots()
sns.heatmap(data.corr(),annot=True)
```

```python
sns.displot(data=data,x='Y house price of unit area',kde=True,color='green')
```

```python
x=data.drop('Y house price of unit area',axis=1)
x
```

```python
y=data['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False)#include bias is used to remose the unwanted noise
poly_fit=polynomial_reg.fit_transform(x)
poly_fit
```

```python
poly_fit.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_fit,y,test_size=0.3,random_state=101)
```

```python
polymodel=LinearRegression()
polymodel.fit(x_train,y_train)
```

```python
y_predict=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y_test':y_test,'y_predict':y_predict})
```

```python
poly_mae=metrics.mean_absolute_error(y_test,y_predict)
poly_mse=metrics.mean_squared_error(y_test,y_predict)
poly_RMSE=np.sqrt(poly_mse)
```
