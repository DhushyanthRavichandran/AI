```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
df
```

```python
features=['Annual Income (k$)','Spending Score (1-100)']
x=df[features]
```

```python
x_stand=StandardScaler().fit_transform(x)
x_stand
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_stand)
clusters
```

```python
centroids=pd.DataFrame(x_stand).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=clusters,cmap='viridis',s=50)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
